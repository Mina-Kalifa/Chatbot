```python
import json
import random
import nltk
from nltk.stem import WordNetLemmatizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD

# load the intents file
with open('football.txt') as file:
    data = json.load(file)

# extract the patterns and responses from the intents
words = []
classes = []
documents = []
for intent in data['intents']:
    for pattern in intent['patterns']:
        # tokenize each word in the pattern
        w = nltk.word_tokenize(pattern)
        words.extend(w)
        # add the pattern to the documents list
        documents.append((w, intent['tag']))
        # add the intent tag to the classes list
        if intent['tag'] not in classes:
            classes.append(intent['tag'])

# lemmatize the words and remove duplicates
words = [WordNetLemmatizer().lemmatize(w.lower()) for w in words if w not in '?']
words = sorted(list(set(words)))
classes = sorted(list(set(classes)))
# create training data
training = []
output_empty = [0] * len(classes)
for doc in documents:
    # initialize bag of words for each pattern
    bag = []
    pattern_words = doc[0]
    pattern_words = [WordNetLemmatizer().lemmatize(word.lower()) for word in pattern_words]
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)
    # create output row for each intent
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    # add the pattern and output row to the training data
    training.append([bag, output_row])

# shuffle the training data
random.shuffle(training)
training = np.array(training)

# split the training data into X and Y
X_train = list(training[:, 0])
Y_train = list(training[:, 1])

# define the neural network model
model = Sequential()
model.add(Dense(128, input_shape=(len(X_train[0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(Y_train[0]), activation='softmax'))

# compile the model
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

# fit the model to the training data
model.fit(np.array(X_train), np.array(Y_train), epochs=200, batch_size=5, verbose=1)

# save the trained model
model.save('football_chatbot_model')
print('Football chatbot model trained and saved.')
```

```python
import numpy as np
from tensorflow.keras.models import load_model

# load the trained model
model = load_model('football_chatbot_model')

# load the words and classes from the JSON file
with open('football.txt') as file:
    data = json.load(file)

# preprocess the user input
def clean_up_sentence(sentence):
    # tokenize the sentence
    sentence_words = nltk.word_tokenize(sentence)
    # lemmatize the words
    sentence_words = [WordNetLemmatizer().lemmatize(word.lower()) for word in sentence_words]
    return sentence_words

# create a bag of words from the preprocessed sentence
def bag_of_words(sentence):
    sentence_words = clean_up_sentence(sentence)
    bag = [0] * len(words)
    for w in sentence_words:
        for i, word in enumerate(words):
            if word == w:
                bag[i] = 1
    return np.array(bag)

# generate a response to the user input
def predict_intent(sentence):
    # generate a bag of words from the user input
    bow = bag_of_words(sentence)
    # predict the intent of the user input
    result = model.predict(np.array([bow]))[0]
    # return the predicted intent tag and probability
    prediction = {'tag': classes[np.argmax(result)], 'probability': str(result[np.argmax(result)])}
    return prediction

# receive input from the user and generate a response
while True:
    message = input('You: ')
    if message.lower() == 'quit':
        break
    else:
        prediction = predict_intent(message)
        if float(prediction['probability']) > 0.5:
            for intent in data['intents']:
                if intent['tag'] == prediction['tag']:
                    response = random.choice(intent['responses'])
            print('Bot:', response)
        else:
            print('Bot: Sorry, I do not understand your question.')
```
